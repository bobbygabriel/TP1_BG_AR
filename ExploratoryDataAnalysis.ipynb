## Frequency and Proportion for Categorical Variables

```python
# Finding the frequency of each category
freq_table = df['Condition'].value_counts()
prop_table = df['Condition'].value_counts(normalize=True)

freq_prop_table = pd.concat([freq_table, prop_table], axis=1)

# Renaming the columns
freq_prop_table.columns = ['Frequency', 'Proportion']

print(freq_prop_table)

```

## Correlation

```python
heat_map_corr=df[['EaseOfUse','Effective','Price', 'Reviews', 'Satisfaction']].corr(method='pearson')
print(heat_map_corr)
cols=['EaseOfUse','Effective','Price', 'Reviews', 'Satisfaction']
heat_map=sns.heatmap(heat_map_corr,
                    annot=True,
                    xticklabels=cols,
                    yticklabels=cols,
                    cmap="Blues")
```

## Anything interesting 

```python
hist_plot=sns.histplot(x='EaseOfUse',data=df)
mean=df['EaseOfUse'].mean()
median=df['EaseOfUse'].median()
# Calculating the mean
plt.title('Ease of Use vs Count')
plt.axvline(mean, color='red')
plt.axvline(median, color='green')
plt.show()
#ploting the mean
```

```python
hist_plot=sns.histplot(x='Effective',data=df)
mean=df['Effective'].mean()
median=df['Effective'].median()
# Calculating the mean
plt.title('Effective vs Count')
plt.axvline(mean, color='red')
plt.axvline(median, color='green')
plt.show()
#ploting the mean
```

```python
scatter_plot=sns.scatterplot(x='Effective',y='Satisfaction',data=df,color='purple')
plt.title('Effectiveness vs Satisfaction')
plt.show()
```
